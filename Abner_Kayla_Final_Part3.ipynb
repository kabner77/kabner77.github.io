## How have certain words been used in wine tasting notes over time?
Each year, the magazine Wine Spectator publishes a list of the "Top 100 Wines" of that year. Rated by professionals, they showcase a variety of wine types and prices, but all score at least a 85 on the 100-point wine rating scale. 

Note about the data: This data was web scraped from [WineSpectator's Top 100 Lists]('https://top100.winespectator.com/lists/') back to 1988.

Type a word into the search box below to see how often a word has been used each Top 100 Year.

```python
# @hidden_cell

# adapted code offered by Jill Naiman
from collections import Counter

years = wine['Top_100_Year'] # possible x values

# using ideas from week 5:
@ipyw.interact(SearchTerm = "Try oak, chocolate, fruity")

def plot_line(SearchTerm):     
    mask = wine['Description'].str.find(str(SearchTerm)) != -1 # search term there?
    new_wine = years[mask]
    plotvals = dict(Counter(new_wine.values))
    yvalues = plotvals.values() # counts of that word occ. by year
    xvalues = plotvals.keys()
    plt.plot(xvalues, yvalues, color='#A72920')
    plt.xlabel('WineSpectator Top 100 Year')
    plt.ylabel('Number of wines with ' + SearchTerm + ' in the desc.')
    plt.rcParams['figure.dpi'] = 125
```

# Some interesting trends

```python
years = wine['Top_100_Year'] # possible x values

rich = wine['Description'].str.find(str('rich')) != -1
ripe = wine['Description'].str.find(str('ripe')) != -1
berry = wine['Description'].str.find(str('berry')) != -1

rich_plot = years[rich]
ripe_plot = years[ripe]
berry_plot = years[berry]

plotvalsr = dict(Counter(rich_plot.values))
yr = plotvalsr.values()
xr = plotvalsr.keys()

plotvalsrr = dict(Counter(ripe_plot.values))
yrr = plotvalsrr.values()
xrr = plotvalsrr.keys()

plotvalsb = dict(Counter(berry_plot.values))
yb = plotvalsb.values()
xb = plotvalsb.keys()

plt.plot(xr, yr, color = 'black', label = 'Rich')
plt.plot(xrr, yrr, color = 'green', label = 'Ripe')
plt.plot(xb, yb, color = 'pink', label='Berry')
plt.legend()
plt.title('A Few Example Words')
plt.ylabel('Number of Wines with Word in Tasting Note')
plt.xlabel('WineSpectator Top 100 Year')
```

_More and more berry, less and less ripe._

## Which wines do amateurs rate the highest?

Another source for wine data is the community-sourced [CellarTracker.com](https://www.cellartracker.com). A forum of sorts for the amateur wine lover, CellarTracker allows users to add wines to their personal "cellar," to keep track of which wines they own and loved (or hated). Users can rate wines on the professional 100 point scale, and provide a personal tasting note.

```python
ctwine = pd.read_csv('cellar_500.csv')
```

The visualization below shows the average score of each type of wine (Red, White, etc.), rated by users on CellarTracker.com. 

```python
scorebycolor = ctwine.groupby('Type')['CScore'].mean()
x_scat = scorebycolor.index
y_scat = scorebycolor.values
scatter = plt.scatter(x_scat, y_scat, color='#A72920')
plt.xticks(rotation=45)
```

_Sweet/dessert Reds seem to be killin' it!_

However, we should note the amount of wines of each type in the dataset!

```python
colors = ctwine['Type'].value_counts()
bar = plt.barh(colors.index, colors.values, color='#A72920')
plt.xticks(rotation=45)
```

As you can see, there are very few (only 1!) sweet/dessert reds in this dataset, and many more Red wines of other types. Therefore, because _one_ dessert red was rated highly, it shifts our view of the dataset.
